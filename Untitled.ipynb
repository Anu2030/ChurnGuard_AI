```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set_style('whitegrid')
plt.rcParams['figure.figsize'] = (8,5)
```

```python
df.groupby('Contract')['Churn'].value_counts(normalize=True).unstack()['Yes'].plot(kind='bar', color='coral')
plt.title('Churn Rate by Contract Type')
plt.ylabel('Churn Rate (%)')
plt.xticks(rotation=0)
plt.show()
```

```python
df.groupby('InternetService')['Churn'].value_counts(normalize=True).unstack()['Yes'].plot(kind='bar', color='steelblue')
plt.title('Churn Rate by Internet Service Type')
plt.ylabel('Churn Rate (%)')
plt.xticks(rotation=0)
plt.show()
```

```python
sns.histplot(data=df, x='tenure', hue='Churn', bins=30, kde=True)
plt.title('Tenure Distribution: Churned vs Retained')
plt.show()
```

```python
sns.heatmap(df[['tenure','MonthlyCharges','TotalCharges']].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
binary_cols = ['Partner', 'Dependents', 'PhoneService', 'PaperlessBilling', 'Churn']
for col in binary_cols:
    df[col] = df[col].map({'Yes': 1, 'No': 0})
```

```python
df['gender'] = df['gender'].map({'Male': 1, 'Female': 0})
```

```python
df.select_dtypes(include='object').columns
```

```python
df = pd.get_dummies(df, drop_first=True)
```
